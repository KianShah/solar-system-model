# Building a scale model of the Solar System (International Edition)
Like with any good notebook, let's start with some imports

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import geodatasets as gd

from typing import List, Dict, Tuple, Optional
```

We're going to start off by defining what our scaled distances will be. We'll use the real distances sourced from [NASA](https://www.jpl.nasa.gov/_edu/pdfs/scaless_reference.pdf) and scale them by the scaling factor we want. Even though the source doesn't say, these are obviously average distances; the real distances vary considerably as all planets are in elliptical orbits.

```python
# Planet distances from the Sun
planet_distances: Dict[str, int] = {
    "Mercury": 5.790e7,
    "Venus": 1.082e8,
    "Earth": 1.496e8,
    "Mars": 2.279e8,
    "Jupiter": 7.786e8,
    "Saturn": 1.434e9,
    "Uranus": 2.873e9,
    "Neptune": 4.495e9,
    "Pluto": 5.906e9
}

scale_factor = 300000

# Distances scaled by the `scale_factor`
scaled_planet_distances: Dict[str, int] = {k: round(v/scale_factor, 1) for k, v in planet_distances.items()}
scaled_planet_distances
```

This chosen scaling factor makes Pluto almost half the circumference of the Earth, which gets us right up to our goal line

Now, let's load up our cities database (Courtesy of https://simplemaps.com/data/world-cities)

```python
cities_df = pd.read_csv("worldcities.csv")
```

Next let's visualize the cities on a map. Since there are thousands of cities, we are going to just take the ones with over a million people in them

```python
cities_df = cities_df[cities_df["population"] >= 1000000]
```

And now let's plot it...

```python
world = gpd.read_file(gd.get_path('naturalearthland'))

def plot_cities(_df: pd.DataFrame):
    geometry = [Point(xy) for xy in zip(_df['lng'], _df['lat'])]
    cities = gpd.GeoDataFrame(_df['city'], geometry=geometry)
    base = world.plot(figsize=(10,10))
    cities.plot(ax=base, marker='o', color='black', markersize=3)
```

```python
plot_cities(cities_df)
```

Note how many cities over 1 million China has. It's going to be fairly likely then that in our best scored model, the Sun will start somewhere there. Amazing! Now, our goal is the following: 

For each candidate sun position, we want to use a greedy strategy to calculate the best possible location for each planet. This best location would be the city with the highest population. Visually, this would be equivalent to drawing a circle around the Sun on the map, and finding the largest population city that falls on that circle line. To get more results, we can imagine that we just need the target city to be +- 1% of the actual scaled distance (i.e. for Mercury consider all cities that are 196 +- 1.96km from the sun position). Then, at the end, we have the total population of all cities that we placed a planet at, and that is our "score". We repeat this for every candidate sun position, and use the maximum score we get as our final answer.

# Step 2: Write the damn code

```python
from geopy.distance import geodesic
```

```python
error_pct = 0.01 # "Thickness" of our circles
```

One optimization I decided to prior to running this; if we sort our cities by population first, when we are looking for the largest city that is within the target distance for each planet, we can stop once we find the first one. We won't have to check every city and then take the largest one. Let's do that now.

```python
cities_df = cities_df.sort_values(by="population", ascending=False)
```

```python
def get_largest_pop_candidate_city(sun_index: int, candidate_sun: pd.Series, target_dist_to_sun: float) -> Optional[int]:
    """
        Returns the city with the largest population that is `target_dist_to_sun` km from the `candidate_sun`
    """
    upper = (1+error_pct)*target_dist_to_sun
    lower = (1-error_pct)*target_dist_to_sun
    sun_latlon = (candidate_sun['lat'], candidate_sun['lng'])

    for j in range(len(cities_df)): # Let's go through each city now to check if this is a valid city to put a planet
        if (j == sun_index):
            continue
        candidate_city = cities_df.iloc[j]
        city_latlon = (candidate_city['lat'], candidate_city['lng'])
        distance_from_sun = geodesic(sun_latlon, city_latlon).kilometers
        if lower <= distance_from_sun <= upper:
            return j
    return None
```

```python
def generate_scale_model_candidates():
    """
        Returns a List of [int, List[int], int] tuples.
        
        The first item in the tuple represents the index of the city in `cities_df` that corresponds to the location of the Sun
        The second item in the tuple represents the indicies of the cities in `cities_df` that correspond to the locations of the planets, in the same order as `planet_distances`
        The third item in the tuple represents the sum of the sizes of all cities in the previous list, plus the size of the city corresponding to the Sun
    """
    res: List[Tuple[int, List[int], int]] = []  
    for i in range(len(cities_df)):
        candidate_sun = cities_df.iloc[i]
        candidate_cities: List[int] = [] # List of city indices
        for planet, target_dist_to_sun in scaled_planet_distances.items():
            highest_population_city_ind = get_largest_pop_candidate_city(i, candidate_sun, target_dist_to_sun)
            if not highest_population_city_ind:
                break
            candidate_cities.append(highest_population_city_ind)
        
        if len(candidate_cities) != len(scaled_planet_distances):
            # In this case, we have failed to find a suitable place to put one of the planets, so let's just continue with another potential sun location
            continue

        total_pop = cities_df.loc[candidate_cities]["population"].sum() + candidate_sun["population"]
        res.append((i, candidate_cities, total_pop))
    return res
```

```python
potential_candidate_locations = generate_scale_model_candidates()
```

We did it! However, it took almost 3 minutes to run on my machine. There's certainly some optimizations we might be able to apply here.

Note that this operation is O(n^2), since `get_largest_pop_candidate_city` is O(n) and we are calling the function `n` times. A good potential optimization would be to reduce the number of calls to `geodesic` by checking each city first to see if the difference in latitude is greater than the target distance, and similarly for longitude. Visually, this is equivalent to drawing a bounding box around your 'Sun' city in which our geodesic circle will be circumscribed within. 

Taking it even one step further, if the difference in latitude was small enough, we know that we need enough of a difference in longitude to get within the minimum bounds of the circle, and vice-versa. This is actually equivalent to circumscribing a smaller bounding box *within* the search circle!

Note that if our list was sorted by the distance to the sun, we could perform the search for candidate cities in O(log n) time using binary search. However, since sorting takes O(nlogn), and we would have to do this for every candidate sun, this turns it into O(n^2logn), which is way worse!


Anyway, for now, let's just move on and get to what we really care about.

```python
models = pd.DataFrame.from_records(potential_candidate_locations, columns=["Sun", "Planets", "Total Pop"])
models.sort_values(by="Total Pop", ascending=False, inplace=True)
models.reset_index(inplace=True)
models
```

Wow, the top result has over 100 million people! We can check out the winner by plotting it. First though, let's verify the results by matching the distances to what we had before:

```python
winner: pd.Series = models.iloc[0]
sun: pd.Series = cities_df.iloc[winner['Sun']]
planets: pd.DataFrame = cities_df.loc[winner['Planets']].reset_index()

sun_latlon = (sun['lat'], sun['lng'])
for i, planet in planets.iterrows():
    city_latlon = (planet['lat'], planet['lng'])
    distance_from_sun = geodesic(sun_latlon, city_latlon).kilometers
    expected_dist_from_sun = list(scaled_planet_distances.values())[i]
    if expected_dist_from_sun > distance_from_sun*(1+error_pct) or expected_dist_from_sun < distance_from_sun*(1-error_pct):
        print("ERROR")
    print(f"Expected: {expected_dist_from_sun} km")
    print(f"Actual: {round(distance_from_sun, 1)} km")
    print()
```

No errors, and these numbers look within the expected error range! Now let's plot it.

```python
winner: pd.Series = models.iloc[0]
sun: pd.Series = cities_df.iloc[winner['Sun']].to_frame().T
planets: pd.DataFrame = cities_df.loc[winner['Planets']]

# Combine them
planets = pd.concat([sun, planets], ignore_index=True)

plot_cities(planets)
```

This sucks, we don't know which planet is which, or even which cities these are! Let's try writing a custom `plot_cities` function

```python
from adjustText import adjust_text

def plot_scale_model(sun: pd.Series, planets: pd.DataFrame):
    """
        Plots the Sun and corresponding Planet cities 
    """
    # Combine them for the plotter
    df = pd.concat([sun.to_frame().T, planets], ignore_index=True)

    geometry = [Point(xy) for xy in zip(df['lng'], df['lat'])]
    cities = gpd.GeoDataFrame(df['city'], geometry=geometry)

    fig, ax = plt.subplots(figsize=(12, 10))
    world.plot(ax=ax, color='lightgray', edgecolor='white')
    cities.plot(ax=ax, color='red', markersize=5, alpha=0.7)

    # Add labels for each point
    texts = []
    texts.append(ax.text(sun['lng'] + 0.5, sun['lat'] + 0.3, f"0: {sun['city']} (Sun)"))
    for i, row in planets.iterrows():
        texts.append(ax.text(row['lng'] + 0.5, row['lat'] + 0.3, f"{i+1}: {row['city']} ({list(scaled_planet_distances)[i]})", fontsize=9))
    adjust_text(texts, ax=ax, arrowprops=dict(arrowstyle='->', color='gray'))

    # Add finishing touches
    plt.title('Scale Model of the Solar System')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()
```

```python
winner: pd.Series = models.iloc[0]
sun: pd.Series = cities_df.iloc[winner['Sun']]
planets: pd.DataFrame = cities_df.loc[winner['Planets']].reset_index()

plot_scale_model(sun, planets)
```

We've used the `adjust-text` library to help with adjusting the text on our plot so as to avoid overcrowding around the 'Sun' city.

Let's take a look at some of our other solutions too:

```python
for i in range(1,5):
    model: pd.Series = models.iloc[i]
    sun: pd.Series = cities_df.iloc[model['Sun']]
    planets: pd.DataFrame = cities_df.loc[model['Planets']].reset_index()

    plot_scale_model(sun, planets)
```

# Part 3: Removing Pluto
Sorry, but Pluto is not a planet. I know we've pretended like it is up until now, but let's get our facts straight. It is not, and **never will be** a planet. 

First we need to change our scales. With Pluto out of the picture, we can use an even larger scale model, at 1:225000 scale. Let's adjust our scaled distances accordingly

```python
scale_factor = 225000
scaled_planet_distances: Dict[str, int] = {k: round(v/scale_factor, 1) for k, v in planet_distances.items() if k != "Pluto"}
scaled_planet_distances
```

Notice that now, Neptune is our furthest away at almost 20000 km (half the circumference of the Earth). Now let's regenerate our scale model candidates

```python
potential_candidate_locations = generate_scale_model_candidates()
```

```python
models = pd.DataFrame.from_records(potential_candidate_locations, columns=["Sun", "Planets", "Total Pop"])
models.sort_values(by="Total Pop", ascending=False, inplace=True)
models.reset_index(inplace=True)
models
```

And now let's plot it...

```python
winner: pd.Series = models.iloc[0]
sun: pd.Series = cities_df.iloc[winner['Sun']]
planets: pd.DataFrame = cities_df.loc[winner['Planets']].reset_index()

plot_scale_model(sun, planets)
```

Cool. It is nice to see North America finally get some recognition, but unfortunately it came at the cost of taking away Africa's. Let's take a look at some of the others.
